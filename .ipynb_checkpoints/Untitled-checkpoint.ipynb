```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import os
from splinter import Browser
from selenium import webdriver
import pandas as pd
```

```python
url = "https://www.indeed.com/jobs?q=&l=virginia"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
jobs = soup.find_all('div', class_=['jobsearch-SerpJobCard', 'unifiedRow', 'row', 'result'])[0].find('a').text.strip()

print(jobs)
```

```python
results = soup.find_all('div', class_="jobsearch-SerpJobCard")
len(results)
print(results)
```

```python
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', class_="turnstileLink").text.strip()
        location = result.find('div', class_="location").text.strip()
        if (title and location):
            print(title)
            print(location)

    except AttributeError as e:
        continue
```
